# 로또 
- [백준 로또 링크](https://www.acmicpc.net/problem/6603)

```python
def backtracking(depth):
    
    for num in lotto:
        if depth == 6:
            print(* ans)

        else:
            if not visited[num]:
                visited[num] = True

                ans.append(num)

                backtracking(depth + 1)
                ans.pop()
                visited[num] = False


while True:
    arr = list(map(int, input().split()))
    
    lotto = arr[1:]
    visited = {num: False for num in lotto}
    ans = []

    # arr[0] == 0일 경우 중단
    if arr[0] == 0:
        break

    else:
        backtracking(0)
```
