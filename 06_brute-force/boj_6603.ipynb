```python
import sys

input = sys.stdin.readline

def backtracking(start, depth):
    if depth == 6:
        print(*ans)
        return

    for i in range(start, len(lotto)):
        ans.append(lotto[i])
        backtracking(i + 1, depth + 1)
        ans.pop()


while True:
    arr = list(map(int, input().split()))

    if arr[0] == 0:
        break

    lotto = arr[1:]
    ans = []

    backtracking(0, 0)
    print()
```
