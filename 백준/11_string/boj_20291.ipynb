```python
from collections import Counter
n = int(input())
extensions = []
d = {}

for _ in range(n):
    name, ext = input().split('.')
    # print(ext)
    extensions.append(ext)

    if ext in d:
        d[ext] += 1
    else:
        d[ext] = 1


for key in sorted(d):
    print(key, d[key])
```
