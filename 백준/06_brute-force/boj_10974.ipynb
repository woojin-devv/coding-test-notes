# 모든 순열 
- [모든 순열 백준 링크](https://www.acmicpc.net/problem/10974)


```python
n = int(input())

visited = [False] * (n+1)
ans = []

def backtracking(depth):
    if depth == n:
        print(* ans)

    for i in range(1, n+1):
        if not visited[i]:
            visited[i] = True
            ans.append(i)

            backtracking(depth + 1)

            ans.pop()
            visited[i] = False


backtracking(0)
```
