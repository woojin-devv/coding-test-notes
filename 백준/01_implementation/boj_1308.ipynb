```python
from collections import Counter

u, n = map(int, input().split())

records = []
prices = []

for _ in range(n):
    name, price = input().split()
    price = int(price)

    records.append((name, price))
    prices.append(price)

counter = Counter(prices)

min_count = min(counter.values())

for p, c in counter.items():
    if c == min_count:
        target_p = p

for name, price in records:
    if price == target_p:
        print(name, price)
        break
```
