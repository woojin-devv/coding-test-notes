# 인간 - 컴퓨터 상호작용


```python
from collections import Counter

w = list(input())
q = int(input())

for _ in range(q):
    a, r, l = input().split()
    r, l = int(r), int(l)

    temp = w[r:l+1]
    freq = Counter(temp)

    print(freq[a])
```
